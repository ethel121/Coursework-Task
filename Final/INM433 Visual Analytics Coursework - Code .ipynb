## Analysing Customer Reviews

We will be using Python to analyse the customer reviews, specifically the columns 'review_comment_title' and 'review_comment_message' because we would like to use a translation function in Python to translate the reviews from Portugese to English. 

```python
import re, nltk
from nltk import word_tokenize
from nltk.corpus import stopwords
from wordcloud import WordCloud, STOPWORDS

#defining stopwords
stop_words = set(stopwords.words('portuguese'))

comments = []

for words in olist_order_reviews_dropna['review_comment_message']:
    only_letters = re.sub("[^a-zA-Z]", " ",words)
    tokens = nltk.word_tokenize(only_letters) #Divide the sentences into smaller parts
    lower_case = [l.lower() for l in tokens] #convert all letters to lower case for consistency
    filtered_result = list(filter(lambda l: l not in stop_words, lower_case))
    comments.append(' '.join(filtered_result))
    
from collections import Counter
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

#Translating the text from Portugese to English
from google_trans_new import google_translator
translator = google_translator()
```

#### Analysing text reviews

```python
# Using CountVectorizer to get the most important word
co_1 = CountVectorizer(ngram_range=(1,1))
counts = co_1.fit_transform(comments)
important_word = pd.DataFrame(counts.sum(axis=0),columns=co_1.get_feature_names()).T.sort_values(0,ascending=False).head(50)

# Next, we reset the index, rename the columns and apply the translate module to get the english translations 
important_word=important_word.reset_index()
important_word.rename(columns={'index':'word',0:'frequency'},inplace=True)

important_word['english_translation'] = important_word['word'].apply(translator.translate)
important_word
```

### Sentiment Analysis

We will be mapping the ratings to 1 = Positive Review for review_score >3 and 0 = Negative Review for review_score<3 and removing neutral reviews (review_score=3).

```python
# Mapping the ratings
olist_order_reviews_dropna['Sentiment_rating'] = np.where(olist_order_reviews_dropna.review_score > 3,1,0)

# Removing neutral reviews 
olist_order_reviews_dropna = olist_order_reviews_dropna[olist_order_reviews_dropna.review_score != 3]

# Printing the counts of each class
olist_order_reviews_dropna['Sentiment_rating'].value_counts()
```

70% of sentiments are classified as positive and 30% of sentiments are classified as negative

```python
olist_order_reviews_dropna.head(10)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
comments =[]

stop_words = set(stopwords.words('portuguese'))

for words in olist_order_reviews_dropna['review_comment_message']:
    only_letters = re.sub("[^a-zA-Z]", " ",words)
    tokens = nltk.word_tokenize(only_letters) #Divide the sentences into smaller parts
    lower_case = [l.lower() for l in tokens] #convert all letters to lower case for consistency
    filtered_result = list(filter(lambda l: l not in stop_words, lower_case))
    comments.append(' '.join(filtered_result))

count_vect = CountVectorizer(stop_words =set(stopwords.words('portuguese')),
                             ngram_range=(1,4)) 

count_vect_comments = count_vect.fit_transform(comments)
count_vect_comments
```

```python
#Split the dataset into training and testing dataset
X_train, X_test, y_train, y_test= train_test_split(count_vect_comments,
                                                                    olist_order_reviews_dropna['Sentiment_rating'],
                                                                    test_size = 0.3,
                                                                    random_state = 42)
```

```python
y_test.value_counts()/y_test.shape[0]
```

69.8% of sentiments are classified as positive and 30.2% of sentiments are classified as negative

#### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
```

```python
# Defining and training the model
lr = LogisticRegression(max_iter=150)
lr.fit(X_train, y_train)

# Predicting the results
test_pred = lr.predict(X_test)

print("F1 score: ", f1_score(y_test,test_pred))
```

```python
#predicted values of Y labels
pred_train = lr.predict(X_train)
pred_test = lr.predict(X_test)

from sklearn.metrics import confusion_matrix

cf = confusion_matrix(y_test,pred_test)

#plot confusion matrix
plt.figure(figsize=(10,10))
sns.heatmap(cf/np.sum(cf), annot=True,fmt='.2%', cmap='Reds')
plt.show()
```
